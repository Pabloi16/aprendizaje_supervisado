# 1. Limpieza de Datos

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/titanic_clean.csv')
```

```python
df.head()
```

```python
print("Valores nulos por columna")
print(df.isnull().sum())
```

```python
df.sample(3)
```

```python
print(f"El shape del conjunto de entrenamiento es {df.shape}. {df.shape[0]} filas y {df.shape[1]} columnas.") 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear mapa de calor de valores nulos
plt.figure(figsize=(10, 6))

sns.heatmap(df.isnull(), 
            cmap="viridis", 
            cbar=False)

plt.title("Mapa de calor de valores nulos del dataset")
plt.xlabel("Columnas")
plt.ylabel("Filas")

plt.show()
```

```python
# Calcular porcentaje de valores faltantes por columna
missing_percentage = (df.isnull().mean() * 100).round(2)

# Ordenar de mayor a menor
missing_percentage = missing_percentage.sort_values(ascending=False)

# Mostrar resultado
print(missing_percentage)
```
